```python
import pandas as pd
```

```python
path = '/Users/jinwon-kim/Desktop/kaggle_data/'
bank_df = pd.read_csv(path+'bank.csv', sep=',')
```

```python
bank_df.head()
```

```python
bank_df.info()
```

```python
# 결손값 제외
bank_df = bank_df.dropna(subset=['job', 'education'])
print(bank_df.shape)
```

```python
# 결손값 보완
bank_df = bank_df.fillna({'concat': 'unknown'})
bank_df.head()
```

```python
bank_df.describe()
```

```python
bank_df = bank_df[bank_df['age'] >= 18]
bank_df = bank_df[bank_df['age'] < 100]
print(bank_df.shape)
```

```python
# 문자열을 수치로 변환
bank_df = bank_df.replace('yes', '1')
bank_df = bank_df.replace('no', '0')

bank_df.head()
```

```python
# 다수의 값을 가지는 데이터 변환
bank_df_job = pd.get_dummies(bank_df['job'])
bank_df_marital = pd.get_dummies(bank_df['marital'])
bank_df_education = pd.get_dummies(bank_df['education'])
bank_df_contact = pd.get_dummies(bank_df['contact'])
bank_df_month = pd.get_dummies(bank_df['month'])
bank_df_job.head()
```

```python
bank_df.columns
```

```python
# 분석 데이터 세트 작성
tmp1 = bank_df[['age', 'default', 'balance', 'housing',
       'loan', 'day', 'duration', 'campaign', 'pdays',
       'previous', 'y']]
tmp1.head()
```

```python
# 더미 변수화한 데이터 결합
tmp2 = pd.concat([tmp1, bank_df_marital], axis=1)
```

```python
tmp3 = pd.concat([tmp1, bank_df_education], axis=1)
tmp4 = pd.concat([tmp1, bank_df_contact], axis=1)
bank_df_new = pd.concat([tmp4, bank_df_month], axis=1)
```

```python
bank_df_new.head()
```

```python
bank_df_new.to_csv('bank-prep.csv', index=False)
```
