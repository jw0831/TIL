### 가지치기를 하는 알고리즘이 apriori 이다  

지지도 support 가 0.1 일경우 마트에서 10건중 1건은 구매한다는 흔하다는 이야기..  
향상도 lift 가 1보다 커야 의미가 있다. 판매를 촉진하는 관점에서  
지지도가 0.01 이상인것만 알아보겠다~ 0.01보다 작으면 의미없는 거래라고 생각하겠다..  
(보안경10/제초기10) / s(제초기 10 / 전체100000) = 1/0.00001 = 10000 이라고 판매량이 높은게 아니어서 지지도는 지정해줘야함.  
지지도가 커도 문제이다. ex) 0.3으로 할 경우.. 이런 상품 자체가 없을수 있다. 적정한 지지도를 찾아야한다  


Transection 에 해당하는  
항목 집합 = {i1, ..., in} = 2의 n승 부분집합-1 (공집합)
2의 n승 -1 개 부분집합으로부터 연관규칙 생성 (임계값 설정)

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
```

```python
dataset = [['Apple', 'Beer', 'Rice', 'Chicken'], 
           ['Apple', 'Beer', 'Rice'], 
           ['Apple', 'Beer'], 
           ['Apple', 'Bananas'], 
           ['Milk', 'Beer', 'Rice', 'Chicken'], 
           ['Milk', 'Beer', 'Rice'], 
           ['Milk', 'Beer'], 
           ['Apple', 'Bananas']]
```

```python
te=TransactionEncoder()
# te_array=te.fit(dataset).transform(dataset)
te_array=te.fit_transform(dataset)
te_array
```

```python
dataset=[['사과','치즈','생수'],
         ['생수','딸기','치즈','고등어'],
         ['수박','사과','생수'],
         ['생수','딸기','치즈','감자']
        ]
```

```python
#데이터프레임으로 변환하시오.
te=TransactionEncoder()
te_array=te.fit_transform(dataset)
te_array
```

```python
df=pd.DataFrame(te_array, columns=te.columns_)
```

```python
freq_itemsets=apriori(df, min_support=0.5, use_colnames=True) #최소 지지도가 0.5 이상인 것들만 추출
freq_itemsets #빈발항목 집합

# ['사과','치즈','생수'],
# ['생수','딸기','치즈','고등어'],
# ['수박','사과','생수'],
# ['생수','딸기','치즈','감자']
#고등어 같은경우는 비빈발 항목이어서 pruning 이 되어버림 (가지치기)
```

```python
from mlxtend.frequent_patterns import association_rules
association_rules(freq_itemsets, metric="confidence", min_threshold=0.3) #연관규칙을 찾는데 사용되는 함수, 규칙이 다 나온다. 빈발항목 집합만 넣어주면 된다.
```

```python
dataset = [['Milk', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'], 
           ['Dill', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'], 
           ['Milk', 'Apple', 'Kidney Beans', 'Eggs'], 
           ['Milk', 'Unicorn', 'Corn', 'Kidney Beans', 'Yogurt'], 
           ['Corn', 'Onion', 'Onion', 'Kidney Beans', 'Ice cream', 'Eggs']]
```

```python
te=TransactionEncoder()
te_array=te.fit_transform(dataset)
te_array
df=pd.DataFrame(te_array, columns=te.columns_)
```

```python
df
```

```python
freq_itemsets=apriori(df, min_support=0.6, use_colnames=True) #최소 지지도가 0.5 이상인 것들만 추출
freq_itemsets 
```

```python
freq_itemsets['length']=freq_itemsets['itemsets'].apply(lambda x: len(x))
```

```python
freq_itemsets
```

```python
#length가 2이면서 support가 0.8 이상인 항목 집합을 모두 출력하시오
freq_itemsets[(freq_itemsets.length>=2) & (freq_itemsets.support>=0.8)]
```

```python
freq_itemsets[freq_itemsets['itemsets']=={'Eggs', 'Onion'}]
```

```python
#거래 데이터를 저장할때 sparse metrics를 사용해서 메모리를 줄일 수 있다.
#Sparse 형태로 표현하여 메모리 낭비 해소
ohtArray=te.fit(dataset).transform(dataset, sparse=True)
ohtArray

sparsedf=pd.DataFrame.sparse.from_spmatrix(ohtArray, columns=te.columns_)
sparsedf
```

### conviction  
conv(X->Y) =   (1-s(Y)) / (1-c(X-Y))

Y 없이 X가 발생한 비율(찾아낸 규칙이 얼마나 잘못되었는지를 확인)

(1-c(X-Y)) = (s(X)-s(X합집합Y)) / s(X)

(1-s(Y)) : Y가 발생하지 않은 비율  
conv(X -> Y) 는 Y가 발생되지 않은 비율이 X가 발생된것 중 Y와 동시에 일어나지 않은 사건의 비율의 몇 배인가를 의미  
ex) 웹문서 
conv(코딩 -> 파이썬): 전체 웹 문서에서 코딩이 없는 비율은 매우 높을 것이다.  
파이썬이 있는 문서에서 코딩이 없는 문서는 그다지 많지 않을 것이다.  
따라서, conv(코딩 -> 파이썬)은 1보다 매우 큰 값을 갖게 될 것이다.  
따라서 코디오가 파이썬의 상관 관계는 매우 높을 것이다.
