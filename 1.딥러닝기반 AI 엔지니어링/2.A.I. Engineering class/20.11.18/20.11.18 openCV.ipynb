### 주의 폴더명에 맞추어서 이미지를 저장해야 한다.

각각의 폴더에서 (example폴더에 총 2장씩 있으면 한꺼번에 가져오겠다 할 경우 batch_size=4) 증식하고자 하는 이미지를 몇개를 가져올 것인가?  
몇개만큼 증식을 시킬 것인가를 따로 코딩이 필요함

```python
import keras
import numpy as np
import cv2
import matplotlib.pyplot as plt
 
np.random.seed(15)
path = 'example'
```

```python
generator = keras.preprocessing.image.ImageDataGenerator(
            rotation_range = 20,
            width_shift_range = 0.2,
            height_shift_range = 0.2,
            rescale = 1. / 255) #(0 < 픽셀값 < 1) 사이로 오게됨
```

```python
batch_size = 4 # 한번에 4장의 이미지를 읽어옴
iterations = 5 # 5번의 이미지 증식
images = []
```

```python
obj = generator.flow_from_directory(
    path,
    target_size = (150, 150),
    batch_size = batch_size,
    class_mode = 'binary') # 0 또는 1로 라벨링이 된다.
```

keras에서는 rgb 방식  
openCV: bgr 방식  
서로 순서가 다르다.
cv2.cvtColor (convert color)  
cv2.cvtColor(img[0], cv2.COLOR_RGB2BGR) -> rgb가 bgr로 변환됨

```python
# 반복해서 처리하는법

for i in enumerate(range(iterations)):
    img, label = obj.next()
    #print(label)
    #print(len(label))
    #print("="*50)
    n_img = len(label)
    #keras에서는 rgb 방식, openCV:bgr 방식 
    base=cv2.cvtColor(img[0], cv2.COLOR_RGB2BGR) # RGB -> BGR 방식으로 변환
    for idx in range(n_img - 1):
        img2 = cv2.cvtColor(img[idx + 1], cv2.COLOR_RGB2BGR)
        base = np.hstack((base, img2))
    images.append(base)   
img = images[0]
for idx in range(len(images) - 1):
    img = np.vstack((img, images[idx + 1]))
```

```python
import numpy
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Flatten
from keras.layers.convolutional import Conv2D
from keras.layers.convolutional import MaxPooling2D
from keras.utils import np_utils
```

```python
train_datagen=ImageDataGenerator(rescale=1./255)
```

```python
train_path="handwriting_shape/train"
```

```python
train_generator=train_datagen.flow_from_directory(
    train_path,
    target_size = (24,24), #이미지 크기가 같거나 작은 값으로 reshape 너무 작으면 중요한 정보를 좀 놓친다 (보통 256,256 / 128,128 정도로 수행한다. 데이터 개수나 시스템 상황에 맞추어서 진행)
    batch_size = 3, # 한 폴더에서 1장씩(random) 이미지를 가져와서 증식 (1장 *3개 폴더 = 3장)
    # 1에폭 = 15batch수행
    class_mode = 'categorical'
)
```

```python
test_datagen=ImageDataGenerator(rescale=1./255)
```

```python
#train과 비슷한 상태의 테스트 이미지
test_path="handwriting_shape/test"
```

```python
test_generator=test_datagen.flow_from_directory(
    test_path,
    target_size = (24,24), #이미지 크기가 같거나 작은 값으로 reshape 너무 작으면 중요한 정보를 좀 놓친다 (보통 256,256 / 128,128 정도로 수행한다. 데이터 개수나 시스템 상황에 맞추어서 진행)
    batch_size = 3, # 한 폴더에서 1장씩(random) 이미지를 가져와서 증식 (1장 *3개 폴더 = 3장)
    # 1에폭 = 15batch수행
    class_mode = 'categorical'
)
```

```python
model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3),
                 activation='relu',
                 input_shape=(24,24,3)))
model.add(Conv2D(64, (3, 3), activation='relu')) #filter = 64개
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(3, activation='softmax'))
```

```python
# (None, 24,24,3) -> conv2d -> (22,22,32) -> 22-3+1 -> (20,20,64) -> (10,10,64) -> (6400) -> (128) -> (3)도형이 3개라서
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```
