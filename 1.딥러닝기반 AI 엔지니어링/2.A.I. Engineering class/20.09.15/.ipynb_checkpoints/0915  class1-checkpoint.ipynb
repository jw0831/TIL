### <U>__배열 연결__</U>: 열 또는 행의 수가 동일한 배열을 연결: stack,  
                    vstack (v=vertical 열의 개수가 동일한: 행이증가),         y축  
                    hstack (h=horizontal 행의 개수가 동일한: 열이 증가), ...  x축  
                    dstack (d=depth 깊이 방향으로 함침) z축

```python
#행의 수가 동일
np.ones((2,3)), np.zeros((2,2))
```

```python
a1=np.ones((2,3))
a2=np.zeros((2,2))
#a1 과 a2는 행의 수가 동일 
# 넘파이 함수임
np.hstack([a1, a2])
```

```python
#열의 수가 동일
np.ones((2,3)), np.zeros((3,3))
```

```python
a1=np.ones((2,3))
a2=np.zeros((3,3))
#a1 과 a2는 열의 수가 동일 
# 넘파이 함수임
np.vstack([a1, a2])
```

```python
#dstack :(depth) 방향으로 배열을 연결
# (3*4), (3*4) => dstack => (3*4*2) #2차원 배열 2개 연결하여 3차원 배열 1개 생성
a1=np.ones((3,4))
a2=np.zeros((3,4))
print(np.dstack([a1, a2]))
np.dstack([a1,a2]).shape #(3,4,2) == (행, 열, 깊이)
```

```python
# stack 함수는 dstack의 기능을 확장, dstack 은 마지막 차원으로 연결
# stack은 사용자가 지정한 축으로 배열을 연결 (axis=0, 디폴트, 가장 앞쪽에 차원이 추가)

a1.shape #3,4
a2.shape #3,4
a=np.stack([a1, a2])
print(a)
print(a.shape) #depth 가 4
```

```python
a1.shape #3,4
a2.shape #3,4
a00=np.stack([a1,a2]) #default: axis=0
print(a00)
print(a00.shape) #depth 가 4
print('='*50)
a01=np.stack([a1,a2], axis=1)
print(a01)
print(a01.shape) #depth 가 4
print('='*50)
a02=np.stack([a1,a2], axis=2)
print(a02)
print(a02.shape) #depth 가 4
print('='*50)
```

```python
#hstack 하고 비슷하지만, 두 배열을 좌우로 연결
np.r_[np.array([1,2,3]), np.array([4,5,6])] #array([1, 2, 3, 4, 5, 6])   1차원


#차원을 증가시킨 후 좌우로 연결
np.c_[np.array([1,2,3]), np.array([4,5,6])] #array([[1, 4],
                                           #        [2, 5],
                                           #        [3, 6]])  2차원

#         밑줄은 인덱서 이지만 대괄호가 붙는 함수를 인덱서 라고 부름
```

```python
# tile: 동일한 배열을 반복해서 연결
a=np.array([[0,1,2], 
            [3,4,5]])
np.tile(a,2)
# array([[0, 1, 2, 0, 1, 2],
#        [3, 4, 5, 3, 4, 5]])

np.tile(a, (3,2))
# 2행 3열의 배열을 -> 타일(3,2) -> 6행6열 배열 생성  (2*3, 3*2) -> (6,6)
# array([[0, 1, 2, 0, 1, 2],
#        [3, 4, 5, 3, 4, 5],
#        [0, 1, 2, 0, 1, 2],
#        [3, 4, 5, 3, 4, 5],
#        [0, 1, 2, 0, 1, 2],
#        [3, 4, 5, 3, 4, 5]])
```

### 핵심  
데이터프레임 인덱스 조작 (설정 / 제거)  
set_index: __기존 행 인덱스를 제거 하고, 데이터 컬럼 중 어느 하나를 인덱스로 설정__  
reset_index: __기존 행 인덱스를 제거하고, 인덱스를 데이터 열로 추가__  
  

```python
myList=list("ABCDE") #['A', 'B', 'C', 'D', 'E'] 요소가 5개

# rand함수: 0<=~<1 사이 의 난수를 생성
myArr=np.round(np.random.rand(3,5),2) #3행 5열 난수 배열 생성 , np.round( ,2) 는 소수이하 2째자리
```

```python
#myList 와 myArr을 연결, 위아래로 연결 (vstack)
np.vstack([myList, myArr])

df1=pd.DataFrame(np.vstack([myList, myArr]).T,
                 columns=['C1','C2','C3','C4']) #(5,4)
df1
```

```python
#set_index 는 데이터프레임에 속함
df2=df1.set_index(['C1'])
df2
```

```python
#df2에서 C2열을 행 인덱스로 지정
df2.set_index(['C2']) #기존 인덱스가 지워짐
```

```python
#df2에서 C2열을 행 인덱스로 지정
#기존 행 인덱스를 제거하고, 인덱스를 데이터 열로 추가 (reset_index)
df2
df2.reset_index()
```

```python
# reset_index를 하는데 다시 못돌아오게 하려면
df2.reset_index(drop=True)
```

```python
#다중인덱스
# 데이터프레임의 columns 인수를 행렬로 주게되면 다중 인덱스
pd.DataFrame(np.random.randn(5,4))

# 열 인덱스 
# 0 1 2 3
# 을
# 아래와 같이 변경
# A         B
# C1   C2   C3   C4
```

```python
#stack (위/아래)
#unstack (좌/우)

# 튜플로 부터 가져온 데이터로 다중인덱스를 만듬
mulIndex=pd.MultiIndex.from_tuples([('A', 'C1'),('A', 'C2'),
                                   ('B', 'C3'),('B', 'C4')])
mulIndex #다중 인덱스를 반드는 법
```

```python
data=pd.DataFrame(np.random.randn(5,4),
                columns=mulIndex)
data
```

```python
df3=pd.DataFrame(np.random.randn(5,4),
                columns=[['A','A','B','B'],
                        ['C1','C2','C3','C4']]) #list 의 list 속성을 주면 = multiIndex
df3
```

```python
df3.columns #다중인덱스 이기때문에 MultiIndex
```

```python
df3.columns.names=['Cidx1','Cidx2']
df3 #각각의 행에대한 이름
```

```python
df4=pd.DataFrame(np.random.randn(6,4),
            index=[['M','M','M','F','F','F'],
                   ['id1','id2','id3','id4','id5','id6']],
            columns=[['A','A','B','B'],
                    ['C','D','C','D']])
df4.index.names=['ridx1','ridx2']
df4.columns.names=['cidx1','cidx2']
```

```python
df4
```

```python
# list comprehension
df4=pd.DataFrame(np.random.randn(6,4),
            index=[['M','M','M','F','F','F'],
                   ["id"+str(i+1) for i in range(3)]*2],
            columns=[['A','A','B','B'],
                    ['C','D','C','D']])
df4.index.names=['ridx1','ridx2']
df4.columns.names=['cidx1','cidx2']
```

```python
df4
```

```python
#stack(): 열 인덱스 -> 행 인덱스 쪽으로 오는데, 반시계 뱡항으로 90도 회전
# vice versa
#stack(): 행 인덱스 -> 열 인덱스 쪽으로 오는데, 시계 뱡항으로 90도 회전

#cidx1에 해당되는 열 인덱스가 (A, B) 행 인덱스로 지정
df4.stack("cidx1") #이름을 주면 이름에 해당하는 index가 이동한다.
```

```python
df4.stack("cidx2")
```

```python
df4.stack(1) #햇갈리니, 이름을 주는게 편하다
```

```python
df4
```

```python
df4.unstack("ridx1") #df4.unstack(0)
```

```python
df4.unstack("ridx2") #df4.unstack(1)
```
