```python
def plot_hist_each_column(df):
    plt.rcParams['figure.figsize']=[20,16]
    fig = plt.figure(1)
    #df열 개수 만큼의 subplotd을 출력 합니다.
    for i in range(len(reg_df.columns)):
        ax=fig.add_subplot(5,5,i+1)
        plt.hist(reg_df[reg_df.columns[i]], bins=50)
        ax.set_title(reg_df.columns[i])
    plt.show()
plot_hist_each_column(reg_df)
```

```python
def standard_scaling(df, scale_columns):
    for col in scale_columns:
        try:
            series_mean = df[col].mean()
            series_std = df[col].std()
            df[col] = df[col].apply(lambda x: (x-series_mean)/series_std)
        except ZeroDivisionError:
            df[col] = df[col].apply(lambda x: (x+1-series_mean)/(series_std+1))
    return df

        
scale_columns=['교육훈련비','난방 공용','수도 공용','인건비','전기 공용','제사무비','제세공과금','경비비','소독비','수선유지비','일반관리비','청소비','공용관리비','승강기유지비','그밖의부대비용']
reg_standard_df=standard_scaling(reg_df, scale_columns)
```
