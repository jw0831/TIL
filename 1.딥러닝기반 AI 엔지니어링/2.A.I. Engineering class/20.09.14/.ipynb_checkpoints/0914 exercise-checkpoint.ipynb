## 1-2. 유클리디안 거리 기반
(4.9, 3.5, 1.4, 0.2) -> 어떤 종?

```python
from scipy.spatial import distance
```

```python
x=pd.DataFrame([{'sepal.length': 4.9,
                 'sepal.width': 3.5,
                 'petal.length': 1.4,
                 'petal.width': 0.2}],
                index=['x'])
```

```python
df=pd.concat([iris,x])
df
```

```python
np.array(df.iloc[:-1,:4])
```

```python
dist_list=[]
for i,j in enumerate (np.array(df.iloc[:,:4])):
#     print(j, i)
#     if i < len(np.array(df.iloc[:,:4])):
    a=np.array(df.iloc[-1:,:4])
    b=j
    eucliDist=distance.euclidean(a,b)
    dist_list.append(eucliDist)
```

```python
df_euc=pd.DataFrame(dist_list, columns=['euclideanDist'])
```

```python
idf=df.join(df_euc)
```

```python
idf.head()
```

```python
idf.sort_values(by='euclideanDist').variety[0]
```
