```python
import pandas as pd
import numpy as np
from keras.layers import *
from keras.models import *
from keras.callbacks import *
from datetime import datetime
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
import datetime as dt
```

```python
df=pd.read_csv("maintenanceCost_stacked.csv", parse_dates=['Date'])
df.head()
```

```python
df.rename(columns={'    가스 공용':'가스 공용','    교육훈련비':'교육훈련비','    그밖의부대비용':'그밖의부대비용',
          '    급탕 공용':'급탕 공용','    난방 공용':'난방 공용', '    수도 공용':'수도 공용','    인건비':'인건비',
          '    전기 공용':'전기 공용', '    제사무비': '제사무비','    제세공과금':'제세공과금', '    차량유지비':'차량유지비', 
           '    피복비':'피복비', '  경비비':'경비비', '  소독비':'소독비', '  수선유지비':'수선유지비',
       '  승강기유지비':'승강기유지비', '  위탁관리수수료':'위탁관리수수료', '  일반관리비':'일반관리비', '  지능형홈네트워크설비유지비':'지능형홈네트워크설비유지비', '  청소비':'청소비',
       '공용관리비':'공용관리비'}, inplace=True)
```

```python
exercise_columns=['Date','난방 공용','수도 공용','인건비','전기 공용','제사무비','제세공과금','경비비','수선유지비','일반관리비','청소비','공용관리비']
```

```python
new_df=df[exercise_columns]
new_df.head()
```

```python
new_df['월']=new_df['Date'].dt.month
```

```python
new_df['년']=new_df['Date'].dt.year
```

```python
new_df.head()
```

```python
scaler=MinMaxScaler(feature_range=(0,1))
```

```python
new_df.columns[1:-2]
```

```python
# new_df.loc[:,"난방 공용":"공용관리비"]
new_df.iloc[:,1:-2].head()
```

```python
scale_df=new_df.copy()
```

```python
# scale_df.iloc[:,1:-2] = scaler.fit_transform(scale_df.iloc[:,1:-2])
scale_df.iloc[:,1:]=scaler.fit_transform(scale_df.iloc[:,1:])
```

```python
scale_df.head()
```

```python
#종속변수를 맨뒤로 보내주기
#종속변수에 속하지 않는 컬럼들을 그대로 넣어주고 마지막에 원하는 종속변수 컬럼을 더해주기
scale_df=scale_df[[c for c in scale_df if c not in ['공용관리비']] + ['공용관리비']]
```

```python
scale_df.head()
```

```python
scale_df=scale_df[[c for c in scale_df if c not in ['월','년']]]
```

```python
scale_df.head()
```

```python
xylist=scale_df.iloc[:,1:].values.tolist()
print(len(xylist))
xylist[12]
```

```python
window_size = 3 #3달 steps 넣어보기로
X=[]
Y=[]
for i in range(len(xylist)-window_size):
    X.append([xylist[i+j] for j in range(window_size)])
    Y.append(xylist[window_size+i][10])
#     print(xylist[window_size+i][12])
```

```python
Y
```

```python
X=np.array(X)
Y=np.array(Y)
```

```python
X.shape # (48개의 샘플, 12번의 timestep, 13개의 피처)
Y.shape
```

```python
# 40개 정도를 트레이닝
# 8개 테스트
train_test_split = 48
# train
X_train = X[:train_test_split,:]
Y_train = Y[:train_test_split]
# test
X_test = X[train_test_split:,:]
Y_test = Y[train_test_split:]
```

```python
X_train.shape, Y_train.shape
```

```python
X_test.shape, Y_test.shape
```

```python
model = Sequential()
model.add(LSTM(1, input_shape=(3,11),return_sequences=False)) #4 는 하이퍼 파라미터
model.add(Dropout(0.2))
model.add(Dense(1, activation='linear')) #output은 1개
model.compile(loss='mse', optimizer='adam')
model.summary()
# model = Sequential()
# model.add(LSTM(1, input_shape=(3,11), return_sequences=False)) #4 는 하이퍼 파라미터
# model.add(Activation('linear')) #output은 1개
# model.compile(loss='mse', optimizer='adam')
# model.summary()
```

```python
hist=model.fit(X_train, Y_train, epochs = 500, batch_size=1, verbose=0)
```

```python
model.save("hist.h5")
```

```python
testpredict=model.predict(X_test)
```
