```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train=pd.read_csv('train.csv',parse_dates=['YearBuilt','YearRemodAdd'])
train
```

```python
test=pd.read_csv('test.csv',parse_dates=['YearBuilt','YearRemodAdd'])
```

```python
train['YearBuilt']=train['YearBuilt'].dt.year
train['YearBuilt']=train['YearBuilt'].astype('category')
train['YearRemodAdd']=train['YearRemodAdd'].dt.year
train['YearRemodAdd']=train['YearRemodAdd'].astype('category')

test['YearBuilt']=test['YearBuilt'].dt.year
test['YearBuilt']=test['YearBuilt'].astype('category')
test['YearRemodAdd']=test['YearRemodAdd'].dt.year
test['YearRemodAdd']=test['YearRemodAdd'].astype('category')
```

```python
train.isnull().sum()[train.isnull().sum()>0]
droplist=['Alley','FireplaceQu','PoolQC','Fence','MiscFeature','LotFrontage']
train=train.drop(droplist,axis=1)
test=test.drop(droplist,axis=1)
```

```python
#작은 결측값 
train.MasVnrArea.fillna(round(train.MasVnrArea.mean()),inplace=True)
train.GarageYrBlt.fillna(round(train.GarageYrBlt.mean()),inplace=True)

test.MSZoning.fillna('RL',inplace=True)
test.Utilities.fillna('NoSeWa',inplace=True)
test.GarageYrBlt.fillna(round(test.GarageYrBlt.mean()),inplace=True)
test.MasVnrArea.fillna(round(test.MasVnrArea.mean()),inplace=True)
test.GarageCars.fillna(round(test.GarageCars.mean()),inplace=True)
test.GarageArea.fillna(round(test.GarageArea.mean()),inplace=True)
test.BsmtFinSF1.fillna(round(test.BsmtFinSF1.mean()),inplace=True)
test.TotalBsmtSF.fillna(round(test.TotalBsmtSF.mean()),inplace=True)
```
