```python
mango=pd.read_excel("mango_df.xlsx")
mango=mango[['가게명','주소','Latitude','Longitude']]
```

```python
dining=pd.read_excel("code_df.xlsx")
dining=dining[['가게명','주소','Latitude','Longitude']]
```

```python
md=pd.concat([mango,dining]).reset_index()
md.drop(columns=['index'], inplace=True)
```

```python
score=pd.read_csv("total_fn.csv")
```

```python
score=score[['가게명','점수_mango','점수_dining','점수_naver']]
score.head()
```

```python
score['all_score']
```

```python
score['dc_score']=score.loc[:,'점수_dining']*5/100
```

```python
score['mean_score']=score.loc[:,['점수_mango','점수_naver','dc_score']].mean(axis=1).round(1)
```

```python
score=score[['가게명','mean_score']]
```

```python
len(score), len(md), len(dist)
```

```python
join1=pd.merge(dist,score, on='가게명') #len = 266
```

```python
join2=pd.merge(join1, md, on='가게명')
```
